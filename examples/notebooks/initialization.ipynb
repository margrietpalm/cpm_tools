# Create randomly dispersed round cells

1. Seed single pixels at a minimum distance of 2*r
2. Grow the seeds into cells
3. Save results to tiff

```python
w = 100
h = 100
ncells = 25
r = 5

# seed pixels
grid = seed_cells(w,h,ncells,pad=2*r,dist=2*r)

# add pixels to each seed to make circles
grid = grow_cells_round(grid,r)

# show results
plt.imshow(grid,cmap='Greys')

# save to tiff
write_to_tiff(grid,'../images/cells_round.tiff')
```

# Create randomly dispersed *realistic* cells

1. Seed single pixels
2. Grow the seeds into *realistic* cells using the diffusion limited aggregation algorithm
3. Save results to tiff

```python
w = 100
h = 100
ncells = 25
volume = 80

# seed pixels
grid = seed_cells(w,h,ncells,pad=5)

# add pixels to each seed to make circles
grid = grow_cells_DLA(grid,volume)

# show results
plt.imshow(grid,cmap='Greys')

# save to tiff
write_to_tiff(grid,'../images/cells_DLA.tiff')
```
